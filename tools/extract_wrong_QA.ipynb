```python
import pandas as pd
import numpy as np
import json
```

```python
val_csv = pd.read_csv('./val.csv')
val_csv['video_qid'] = val_csv["video"].astype('string') + "_" +val_csv['qid'].astype('string')
```

```python
with open('./HGA-bert-ft-h256-val.json') as f:
    val_result = json.load(f)
    
wrong_result={}
for id in val_result:
    if val_result[id]['prediction'] != val_result[id]['answer']:
        wrong_result[id] = val_result[id]
wrong_result
```

```python
wrong_csv = pd.DataFrame()
for idx in wrong_result:
    tmp = val_csv.loc[val_csv['video_qid']==idx, 
                             ['video_qid','type','question','answer','a0','a1','a2','a3','a4']]
    tmp.insert(4,'pred_answer', wrong_result[idx]['prediction'])
    wrong_csv = wrong_csv.append(tmp)
wrong_csv
```
